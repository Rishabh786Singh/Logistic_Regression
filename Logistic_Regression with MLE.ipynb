```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import random
from sklearn.utils import shuffle
from IPython.display import clear_output

%matplotlib inline
```

```python
data=pd.read_csv('voice.csv')
data.head(10)
```

```python
data.isna().sum()
```

```python
li=list(set(data['label']))
print(li)
data['label']=data['label'].replace(li,[0,1])
```

```python
data=shuffle(data)
data.head(10)
```

```python
y=np.array(data['label']).reshape(1,len(data['label']))
x=data.drop(['label'],axis=1)
x=x/x.max()
x=np.array(x).reshape(3168,20)
print(x.shape,'\n', y.shape)
```

```python
def sigmoid(m,x,c):
    return 1/(np.exp(-(np.dot(m,x.T)+c))+1)

m=np.random.randn(1,20)
c=random.random()
#print(sigmoid(m,x,c).shape)

def cost(y_pre,y):
    #print(x[0].shape)
    y_pred=[]
    for i in range(len(y.T)):# if y_actual==1 then cost function is used -log(y_predicted)
        if(y[0][i]==1):
            y_pred.append(-np.log(y_pre[0][i]))
            #print(y[0][i]," ",y_pre[0][i])
                          
        else:        # if y_actual==0 then cost function is used -log(1-y_predicted)
            #print(y_pre[0][i])
            y_pred.append(-np.log(1-y_pre[0][i]))
            
    res=np.mean(y_pred)
    #print(np.array(y_pred).shape)
    return res

def der_slope(m,x,c,y):
    return np.array(np.mean((sigmoid(m,x,c)-y)*x.T, axis=1)).reshape(1,20)

#print(der_slope(m,x,c,y).shape)
         
def inter(m,x,c,y):
    return np.mean(sigmoid(m,x,c)-y)

def accuracy(m,x,c,y):
    y_predicted=sigmoid(m,x,c)
    #print(y_predicted.shape," ",y.shape)
    y_predicted[y_predicted>=0.5]=1
    y_predicted[y_predicted<0.5]=0
    correct=len(y_predicted[y_predicted==y])
    #print(correct," ",len(y.T))
    return (correct/len(y.T))*100

#print(accuracy(m,x,c,y))
#print(cost(sigmoid(m,x,c),y))
```

```python
#now lets check out our good work
m=np.random.randn(1,20)#randomly initialize value of m with the help of normal distribution
#print(m)
c=random.random()
alpha=10#learning rate
cost1=[]
accu=[]
for i in range(0,300):
    
    cost1.append(cost(sigmoid(m,x,c),y))#append current error in cost, just to plot and visualize it
    accu.append(accuracy(m,x,c,y))
    m=m-alpha*(der_slope(m,x,c,y))#m=m-learning_rate*d(error)/d(m)
    c=c-alpha*inter(m,x,c,y)#c=c-learning_rate*d(error)/d(c)
    #print(m.shape)
    clear_output(True)#just to remove the previous printed output from the display
    print(i,'  Current error-->',cost(sigmoid(m,x,c),y),"  ",'Current Accuracy--->', accuracy(m,x,c,y))
```

```python
plt.plot(cost1)
plt.xlabel("number of iterations")
plt.ylabel("error")
plt.title("alpha=0.1")
```
